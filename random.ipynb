```python
import os
import pandas as pd

FOLDER_PATH = "tutors_data"   # adjust if needed


def process_file(file_path):
    print(f"Processing: {file_path}")

    df = pd.read_csv(file_path)

    # Make sure badge column exists
    if "badge" not in df.columns:
        print("  -> No badge column found. Skipping.")
        return

    # Normalize badge column to string
    df["badge"] = df["badge"].fillna("").astype(str).str.lower()

    # Create new boolean columns
    df["is_professional"] = df["badge"].str.contains("professional")
    df["is_super_tutor"] = df["badge"].str.contains("super")

    # Save back to same file
    df.to_csv(file_path, index=False)

    print("  -> Updated successfully.\n")


def process_all_files():
    for filename in os.listdir(FOLDER_PATH):
        if filename.endswith(".csv"):
            file_path = os.path.join(FOLDER_PATH, filename)
            process_file(file_path)


if __name__ == "__main__":
    process_all_files()
```
